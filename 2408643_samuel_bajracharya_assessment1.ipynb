##Task - 3 - Outlier Detection:
###Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
# For score column

# finding the lower and upper quartile
sa_score_Q1 = sa_score.quantile(0.25)
sa_score_Q3 = sa_score.quantile(0.75)

# finding the inter quartile range
sa_score_IQR = sa_score_Q3 - sa_score_Q1

# finding the upper and lower bound outliers
sa_lower_score_outlier = sa_score_Q1 - 1.5 * sa_score_IQR
sa_upper_score_outlier = sa_score_Q3 + 1.5 * sa_score_IQR

# creating a dataset of outliers
sa_score_outlier_dataset = south_asian_dataframe[(sa_score < sa_lower_score_outlier) | (sa_score > sa_upper_score_outlier)]
sa_score_outlier_dataset.head()
```

```python
# For GDP column

# finding the lower and upper quartile
sa_GDP_Q1 = sa_GDP.quantile(0.25)
sa_GDP_Q3 = sa_GDP.quantile(0.75)

# finding the inter quartile range
sa_GDP_IQR = sa_GDP_Q3 - sa_GDP_Q1

# finding the upper and lower bound outliers
sa_lower_GDP_outlier = sa_GDP_Q1 - 1.5 * sa_GDP_IQR
sa_upper_GDP_outlier = sa_GDP_Q3 + 1.5 * sa_GDP_IQR

# creating a dataset of outliers
sa_GDP_outliers_dataset = south_asian_dataframe[(sa_GDP < sa_lower_GDP_outlier) | (sa_GDP > sa_upper_GDP_outlier)]
sa_GDP_outliers_dataset.head()
```

```python
plt.figure(figsize=(10, 6))

# Scatter plot for normal points
plt.scatter(
    sa_GDP[(sa_GDP >= sa_lower_GDP_outlier) & (sa_GDP <= sa_upper_GDP_outlier) & (sa_score >= sa_lower_score_outlier) & (sa_score <= sa_upper_score_outlier)],
    sa_score[(sa_GDP >= sa_lower_GDP_outlier) & (sa_GDP <= sa_upper_GDP_outlier) & (sa_score >= sa_lower_score_outlier) & (sa_score <= sa_upper_score_outlier)],
    color = 'blue', label = 'Normal'
)

# Scatter plot for lower outliers
plt.scatter(
    sa_GDP[(sa_GDP < sa_lower_GDP_outlier) | (sa_score < sa_lower_score_outlier)],
    sa_score[(sa_GDP < sa_lower_GDP_outlier) | (sa_score < sa_lower_score_outlier)],
    color = 'red', label = 'Lower Outliers'
)

# Scatter plot for upper outliers
plt.scatter(
    sa_GDP[(sa_GDP > sa_upper_GDP_outlier) | (sa_score > sa_upper_score_outlier)],
    sa_score[(sa_GDP > sa_upper_GDP_outlier) | (sa_score > sa_upper_score_outlier)],
    color = 'green', label = 'Upper Outliers'
)

plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.title('Scatterplot for Upper and Lower Outliers using Score and GDP')

# Add legend
plt.legend()
plt.grid()
plt.show()
```

#Task - 4 - Exploring Trends Across Metrics:
##Tasks:
1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
{pearson correlation} with the Score for South Asian countries.
2. Create scatter plots with trendlines for these metrics against the Score.
3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
South Asian countries.

```python
# function to calculate Pearson correlation
def pearson_correlation(metric, score):
  '''
  This function calculates the pearson correlation coeeficient between any metric and the happiness score.

  Params:
  metric: Metric column passed by the user
  score: Happiness score column passed by the user

  Returns:
  correlation: Pearson correlation coefficient between metric and scorea
  '''
  # calculating mean
  mean_metric = metric.mean()
  mean_score = score.mean()

  # calculating numerator and denominator to find the Pearson correlation
  numerator = sum((metric - mean_metric) * (score - mean_score))
  denominator = np.sqrt(sum((metric - mean_metric)**2) * sum((score - mean_score)**2))

  # calculating correlation
  correlation = numerator / denominator
  return correlation

# calling the function using freedom and generosity columns
freedom_correlation = pearson_correlation(sa_Freedom, sa_score)
generosity_correlation = pearson_correlation(sa_generosity, sa_score)

# printing the correlation
print("The Pearson correlation between Freedom to make life choices and score is: ", freedom_correlation)
print("The Pearson correlation between Generosity and score is: ", generosity_correlation)
```

```python
# scatterplot of two metrics with trendlines

plt.figure(figsize = (14, 6))

# subplot index 2 for scatterplot of freedom to make life choices
plt.subplot(1, 2, 1)
# using regplot to create a scatterplot with a trendline
sns.regplot(x = sa_Freedom, y = sa_score, ci = None, line_kws = {'color': 'lime'}, scatter_kws = {'color': 'purple'})

plt.xlabel('Freedom to make life choices')
plt.ylabel('Score')
plt.title('Scatterplot of Freedom to make life choices against score with trendline.')
plt.grid()


# subplot index 2 for scatterplot of generosity
plt.subplot(1, 2, 2)
# using regplot to create a scatterplot with a trendline
sns.regplot(x = sa_generosity, y = sa_score, ci = None, line_kws = {'color': 'red'}, scatter_kws = {'color': 'turquoise'})

plt.xlabel('Generosity')
plt.ylabel('Score')
plt.title('Scatterplot of Generosity against score with trendline.')

plt.grid()
plt.show()
```

Comparing the scatter plots side by side, we can interpret for both the graphs that the trendlines slope toward the upwards direction which means that there is a strong positive relationship with score in both the cases. The interpretation that we can make from the graphs is that as the value for freedom to make choices or the value for Generosity increases, the value of score increases with them. Comparing the value of correlation coefficient of Freedom to make life choices and Generoisity, the values denote that there is a strong positive relationship of them with score.

###6. Outlier Detection:
• Identify outlier countries in both regions based on Score and GDP per Capita.

• Plot these outliers and discuss their implications.

```python
# For score column of Middle East

# finding the lower and upper quartile
me_score_Q1 = me_score.quantile(0.25)
me_score_Q3 = me_score.quantile(0.75)

# finding the inter quartile range
me_score_IQR = me_score_Q3 - me_score_Q1

# findint the lower and upper bound outliers
me_lower_score_outlier = me_score_Q1 - 1.5 * me_score_IQR
me_upper_score_outlier = me_score_Q3 + 1.5 * me_score_IQR

# creating a dataset of all the outliers
me_score_outlier_dataset = middle_east_dataframe[(me_score < me_lower_score_outlier) | (me_score > me_upper_score_outlier)]
me_score_outlier_dataset.head()
```

```python
# For GDP column of Middle East

# finding the lower and upper quartile
me_GDP_Q1 = me_GDP.quantile(0.25)
me_GDP_Q3 = me_GDP.quantile(0.75)

# finding the inter quartile range
me_GDP_IQR = me_GDP_Q3 - me_GDP_Q1

# findint the lower and upper bound outliers
me_lower_GDP_outlier = me_GDP_Q1 - 1.5 * me_GDP_IQR
me_upper_GDP_outlier = me_GDP_Q3 + 1.5 * me_GDP_IQR

# creating a dataset of all the outliers
me_GDP_outliers_dataset = middle_east_dataframe[(me_GDP < me_lower_GDP_outlier) | (me_GDP > me_upper_GDP_outlier)]
me_GDP_outliers_dataset.head()

```

```python
plt.figure(figsize=(10, 6))

# Scatter plot for normal points
plt.scatter(
    me_GDP[(me_GDP >= me_lower_GDP_outlier) & (me_GDP <= me_upper_GDP_outlier) & (me_score >= me_lower_score_outlier) & (me_score <= me_upper_score_outlier)],
    me_score[(me_GDP >= me_lower_GDP_outlier) & (me_GDP <= me_upper_GDP_outlier) & (me_score >= me_lower_score_outlier) & (me_score <= me_upper_score_outlier)],
    color = 'turquoise', label = 'Normal'
)

# Scatter plot for lower outliers
plt.scatter(
    me_GDP[(me_GDP < me_lower_GDP_outlier) | (me_score < me_lower_score_outlier)],
    me_score[(me_GDP < me_lower_GDP_outlier) | (me_score < me_lower_score_outlier)],
    color = 'maroon', label = 'Lower Outliers'
)

# Scatter plot for upper outliers
plt.scatter(
    me_GDP[(me_GDP > me_upper_GDP_outlier) | (me_score > me_upper_score_outlier)],
    me_score[(me_GDP > me_upper_GDP_outlier) | (me_score > me_upper_score_outlier)],
    color = 'orange', label = 'Upper Outliers'
)

plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.title('Scatterplot for Upper and Lower Outliers using Score and GDP of Middle East')

# Add legend
plt.legend()
plt.grid()
plt.show()
```

We can observe that all the dots are of same color which is for normal outlier. This means that all the scores of the middle eastern countries lie within the bound for outliers indicating the absence of any outlier in the Middle east.
